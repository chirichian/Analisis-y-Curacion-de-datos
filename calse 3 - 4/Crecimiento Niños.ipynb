# Cargamos datos de altura

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
l_age = pd.read_csv ("https://docs.google.com/spreadsheets/d/e/2PACX-1vTVRozth5OL3JRm_X9wBKj7j5XZAbAeG0sBBIKWfYCMX7MS_qiAdcgyTWk1eWNoBq5sCWYNN-E1fKf7/pub?gid=1417284885&single=true&output=csv")
l_age.describe()
```

```python
l_age.dtypes
```

```python
l_age_LE=l_age[["Sex","Agemos","L","M","S","P3","P5","P10","P25","P50","P75","P90","P95","P97"]]

l_age_LE
```

# Conversion de datos

```python
l_age_LE=l_age_LE.apply(pd.to_numeric, errors="coerce")
l_age_LE.dtypes
```

```python
#Seleccionamos los datos de los varones
l_age_LE_boys= l_age_LE[l_age_LE.Sex==1]

#validamos la division
l_age_LE_boys.head
```

```python
fig = plt.figure(figsize=(8,10))
for column in l_age_LE_boys.columns[5:]:
    plt.plot(l_age_LE_boys.Agemos, l_age_LE_boys[column]);

plt.axis(xmin=0)
plt.title("Comparación altura x edad")
plt.ylabel("Kg")
plt.xlabel("Mes")
plt.legend()
plt.show()
```

```python
#Seleccionamos los datos de las chicas
l_age_LE_girls= l_age_LE[l_age_LE.Sex==1]

#validamos la division
l_age_LE_girls.head
```

```python
fig = plt.figure(figsize=(8,10))
for column in l_age_LE_girls.columns[5:]:
    plt.plot(l_age_LE_girls.Agemos, l_age_LE_girls[column]);

plt.axis(xmin=0)
plt.title("Comparación altura x edad")
plt.ylabel("Kg")
plt.xlabel("Mes")
plt.legend()
plt.show()
```

# Cargamos los datos de pesos 

```python
#leemos y cargamos datos en w_age
w_age= pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSBO0i3U2TeyRTDU0MrwLCwncRKk2meg8upW7EtdVPgeMlzmYvQyOHAePHyVl5_x9MDE2sDBs-1CM73/pub?gid=1984889602&single=true&output=csv")

#listamos primeros 10
w_age.head(10)
```

```python
#validamos el tipo de datos
w_age.dtypes
```

```python
#Para realizar los graficos primero dividiremos los datos entre niños y niñas
#Realizamos la primer division
w_age_boys= w_age[w_age.Sex==1]

#validamos la division
w_age_boys.head
```

```python
fig = plt.figure(figsize=(8,10))
for column in w_age_boys.columns[5:]:
    plt.plot(w_age_boys.Agemos, w_age_boys[column]);

plt.axis(xmin=0)
plt.title("Comparación pesos x edad")
plt.ylabel("Kg")
plt.xlabel("Mes")
plt.legend()
plt.show()
```

```python
#Realizamos lo mismo pero con niñas
w_age_girls= w_age[w_age.Sex==2]
w_age_girls=w_age_girls.sort_values(by="Agemos")

w_age_girls
```

```python
fig = plt.figure(figsize=(8,10))
for column in w_age_girls.columns[5:]:
    plt.plot(w_age_girls.Agemos, w_age_girls[column]);

plt.axis(xmin=0)
plt.title("Comparación pesos x edad")
plt.ylabel("Kg")
plt.xlabel("Mes")
plt.legend()
plt.show()
```
